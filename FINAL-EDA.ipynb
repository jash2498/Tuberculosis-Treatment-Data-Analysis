## FAILURE and SMOKING

```python
sns.countplot(df['failure'], hue=df['smoking'])
```

## There is higher likelihood of smoking in the  failure cases

```python
import seaborn as sns
import matplotlib.pyplot as plt


crosstab = pd.crosstab(df['failure'], df['smoking'])

plt.figure(figsize=(8, 6))
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt='d')
plt.title('Correlation between Failure and Smoking')
plt.xlabel('Smoking')
plt.ylabel('Failure')

plt.show()
```

## from above heatmap we can observe clearly that there is a noticeable and statistically significant positive correlation between the 'failure' variable and 'smoking' behavior, suggesting that a higher incidence of failure is associated with a higher likelihood of smoking.
